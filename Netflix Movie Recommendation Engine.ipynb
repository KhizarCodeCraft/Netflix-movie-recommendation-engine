# Data Cleaning

```python
import ast   #load ast(abstract syntax tree) for convert list into string 
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

# Drop Na Row

```python
movies.dropna(inplace=True)
```

## convert list of dictionary into list

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

##  Get Top 3 crew name

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

## Get director name from crew

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
#handle space between words
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# convert overview string into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

## create new column Tags

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(1)
```

## Drop column Genres,keywords,cast and crew

```python
new_movies = movies.drop(columns=['overview','keywords','crew'])
```

```python
new_movies.head(1)
```

## convert tags(list) into strings

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

## import countvectorizer class from sklearn library to convert text into vector

```python
from sklearn.feature_extraction.text import CountVectorizer
#get 5000 most frequent word used in tags with stop_words english
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# list of most frequent 5000 words repeat in tags
cv.get_feature_names()
```

```python
#convert features names into numpy vetor array
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors.shape
```

```python
#import cosine_similarity class to calculate distance between vector
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
# distance of each movie with other movie
similarity
```

```python
#get index of movie on movie name
new_movies[new_movies['title'] == 'The Lego Movie'].index[0]
```

## Functiuon return 5 related movie to given movie name on basis of content based filterinng

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
        
    
```
